```python
import re

def parse_pseudocigar(PseudoCIGAR, ASV):
    
    reference = list(ASV)  
    query = list(ASV)      
    mutations = re.findall(r'(\d+)([ID])=([ACTG])|(\d+)([ACTG])', PseudoCIGAR)
    offset = 0

    for match in mutations:
        if match[1]: 
            position = int(match[0]) - 1 + offset  
            mutation_type = match[1]
            base = match[2]
            
            if position < len(reference): 
                if mutation_type == 'I':  
                    query.insert(position, base)
                    reference.insert(position, '-')
                    offset += 1
                elif mutation_type == 'D':  
                    reference.insert(position, base)
                    query.insert(position, '-')
                    offset += 1
            else:
                print(f"Position {position} out of bounds for reference/query length.")

        elif match[4]:  
            position = int(match[3]) - 1 + offset  
            base = match[4]
            
            if position < len(reference):
                reference[position] = base
            else:
                print(f"SNP Position {position} out of bounds for reference.")

    return ''.join(reference), ''.join(query)


joined[['Reference', 'ASV1']] = joined.apply(lambda row: pd.Series(parse_pseudocigar(row['PseudoCIGAR'], row['ASV'])), axis=1)

```

The reference is the base sequence 
The query is the sequence that contains all the mutations described in the PseudoCIGAR
